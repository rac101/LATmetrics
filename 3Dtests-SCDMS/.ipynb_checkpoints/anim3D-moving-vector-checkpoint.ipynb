Output stereo views to see a 3D view of the vector in the coordinate axes.

```python
import matplotlib.pyplot as plt
from numpy import sin, cos, radians, sqrt, linspace, exp, pi, append, array
%matplotlib inline
import JSAnimation
from JSAnimation import IPython_display
from matplotlib.animation import FuncAnimation
```

```python
a = 70.0         # a = angle between Z-axis and line-of-sight (degrees)
b = 10.0         # b = angle between east and projection of Z-axis on sky (degrees)
sina = sin(a)
sinb = sin(b)
cosa = cos(a)
cosb = cos(b)

delta = 5.0     # delta is the angle between your eyes for the stereo projection (degrees)
delta = radians(delta)
sindelta = sin(delta)
cosdelta = cos(delta)
```

```python
# do the vector axes

su1 = array([0,1,0,0,0,0,0])
sv1 = array([0,0,0,1,0,0,0])
sw1 = array([0,0,0,0,0,1,0])

# project onto the viewing x,y,z axes

x = su1*cosa*cosb+sv1*sinb-sw1*sina*cosb
y =-su1*cosa*sinb+sv1*cosb+sw1*sina*sinb
z = su1*sina+sw1*cosa

# project onto the viewing x',y',z' axes

xd = x*cosdelta+z*sindelta
yd = y
zd =-x*sindelta+z*cosdelta
```

```python
fig, (sp1, sp2) = plt.subplots(1, 2, subplot_kw=dict(frame_on=False),figsize=(8,8))
eye1, = sp1.plot([],[])
eye2, = sp2.plot([],[])
sp1.axis('off')
sp2.axis('off')
sp1.set_xlim(-4,4)
sp2.set_xlim(-4,4)
sp1.set_ylim(-7,7)
sp2.set_ylim(-7,7)
#plt.show()
```

```python
def update(i):
    
# sweep through viewing angles
    a = radians(i)
    sina = sin(a)
    cosa = cos(a)

# draw the axes
    
    su1 = array([0,1,0,0,0,0,0])
    sv1 = array([0,0,0,1,0,0,0])
    sw1 = array([0,0,0,0,0,1,0])

# project onto the viewing x,y,z axes

    x = su1*cosa*cosb+sv1*sinb-sw1*sina*cosb
    y =-su1*cosa*sinb+sv1*cosb+sw1*sina*sinb
    z = su1*sina+sw1*cosa

# project onto the viewing x',y',z' axes

    xd = x*cosdelta+z*sindelta
    yd = y
    zd =-x*sindelta+z*cosdelta
        
    eye1.set_data(x, y)
    eye2.set_data(xd, y)

    return eye1,eye2,
                
FuncAnimation(fig, update, frames=359, interval=50)
```
