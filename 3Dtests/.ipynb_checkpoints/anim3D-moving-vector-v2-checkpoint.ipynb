Output stereo views to see a 3D view of the vector in the coordinate axes.

default axis projection: Z out of screen, Y up, X to the right
desired axis projection: Z up, X a bit to the left, Y a bit to the right. The viewpoint in the +Z hemisphere.
consider an intermediate (semi-desired) desired projection: Z up, X out of the screen, Y to the right.
Transformation from default to semi-desired: 2 rotations involved: -90 degrees around X, then -90 degrees around Z.
Alternatively, axis-relabeling: Z -> X, Y -> Z, X -> Y:  (X,Y,Z) -> (Y,Z,X)

```python
import matplotlib.pyplot as plt
from numpy import sin, cos, radians, sqrt, linspace, exp, pi, append, array
%matplotlib inline
import JSAnimation
from JSAnimation import IPython_display
from matplotlib.animation import FuncAnimation
```

```python
a = 0.0         # a = angle between Z-axis and line-of-sight (degrees)
b = 0.0         # b = angle between east and projection of Z-axis on sky (degrees)
sina = sin(a)
sinb = sin(b)
cosa = cos(a)
cosb = cos(b)

srot1 = 30.0   # rotation1 of axes about Z
srot2 = 15.0   # rotation2 of axes about X

delta = 5.0     # delta is the angle between your eyes for the stereo projection (degrees)
delta = radians(delta)
sindelta = sin(delta)
cosdelta = cos(delta)
```

```python
# do the vector axes

ax = array([0,1,0,0,0,0,0])
ay = array([0,0,0,2,0,0,0])
az = array([0,0,0,0,0,3,0])

# project onto the viewing x,y,z axes

x = ax
y = ay
z = az

# project onto the viewing x',y',z' axes

xd = x*cosdelta+y*sindelta
yd =-x*sindelta+y*cosdelta
zd = z
```

```python
fig, (sp1, sp2) = plt.subplots(1, 2, subplot_kw=dict(frame_on=False),figsize=(8,8))
eye1, = sp1.plot([],[])
eye2, = sp2.plot([],[])
sp1.axis('off')
sp2.axis('off')
sp1.set_xlim(-4,4)
sp2.set_xlim(-4,4)
sp1.set_ylim(-7,7)
sp2.set_ylim(-7,7)
plt.show()
```

```python
def update(i):
    
# sweep through viewing angles
    b = radians(i)
    sinb = sin(b)
    cosb = cos(b)

# draw the axes
    
    ax = array([0,1,0,0,0,0,0])
    ay = array([0,0,0,2,0,0,0])
    az = array([0,0,0,0,0,3,0])

# project onto the viewing x,y,z axes

    x = ax
    y = ay
    z = az

# project onto the viewing x',y',z' axes

    xd = x*cosdelta+y*sindelta
    yd =-x*sindelta+y*cosdelta
    zd = z
    
    eye1.set_data(x, y, z)
    eye2.set_data(xd, yd, zd)

    return eye1,eye2,
                
FuncAnimation(fig, update, frames=359, interval=50)
```
