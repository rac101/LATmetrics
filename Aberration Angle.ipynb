The aberration equation is:

tan(phi) = sin(theta)/(cos(theta) - v/c)

Where

phi = apparent, aberrated angle from moving observer to celestial object

theta = true, un-aberrated angle from moving observer to celestial object

```python
import numpy as np
#from math import *
%matplotlib notebook
import matplotlib.pyplot as plt, mpld3
```

```python
v = 7.6   # km/s, for FGST in low-earth orbit
c = 299792.458  # km/s, in vacuum

fig, ax = plt.subplots()
theta = np.linspace(-999, 999, 1999)
theta *= 0.1
st = np.sin(theta*np.pi/180)
ct = np.cos(theta*np.pi/180)
#plt.plot (theta,st)
#plt.plot (theta,ct)
#tanphi = st/(ct - v/c)
phi = np.degrees(np.arctan2(st, ct - v/c))
dang = (theta - phi)*3600
ax.plot(theta,dang)

x_ticks = np.arange(-90, 91, 30)                                                                                            
ax.set_xticks(x_ticks)
y_ticks = np.arange(-6, 7, 1)                                                                                            
ax.set_yticks(y_ticks)
ax.grid('on')
ax.set_xlabel('True Angle from Observer to Object (degrees)')
ax.set_ylabel('Aberration (arcseonds)')
```
