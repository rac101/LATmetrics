```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import JSAnimation
from JSAnimation import IPython_display
from matplotlib.animation import FuncAnimation
```

```python
# define vector of rotation axis (r), and rotation angle (theta)

r = (-1.0,1.0,1.0)/sqrt(3.0)
theta = 45.0
rtheta = np.radians(theta)
```

```python
# define rotation matrix, m

# r = (u,v,w)
u = r[0]
v = r[1]
w = r[2]
ct = np.cos(rtheta)
st = np.sin(rtheta)
tt = 1.0 - ct

m = [[tt*u*u-ct, tt*u*v-st*w, tt*u*w+st*v],[tt*u*v+st*w, tt*v*v+ct, tt*v*w-st*u],[tt*u*w-st*v, tt*v*w+st*u, tt*w*w+ct]]
```

```python
x = np.matrix([[1],[0],[0]])
y = np.matrix([[0],[1],[0]])
z = np.matrix([[0],[0],[1]])
mx = m@x
my = m@y
mz = m@z
mx , my, mz
```
