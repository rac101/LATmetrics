```python
import numpy as np
%pylab inline
rcParams['figure.figsize'] = 12, 8
rcParams['font.size'] = 16
#import sys
#from subprocess import Popen, PIPE, STDOUT
from math import asin
#import operator
#from functools import reduce
sys.path.append('/Users/rc/Documents/GLAST/sgp4-1.4')
from sgp4.earth_gravity import wgs84
from sgp4.io import twoline2rv
from sgp4.ext import jday, invjday, days2mdhms
```

```python
satellite = twoline2rv(tle[0], tle[1], wgs84)
```

```python
oformat = "%4d %02d %02d %02d %02d %02d %12.7f %12.6f %12.6f %12.6f %8.3f\n"
fo = open("beta.txt", "w")
ws = tle[0].split()
ydoy = float(ws[3])
year = int(ydoy/1000) + 2000
jd0 = jday(year,1,1,0,0,0)
ylen = 365
if year % 4 == 0:
    ylen = 366
fdoypl = []
bpl = []
for doy in range(ylen):
    for hour in range(24):
# add a small amount to the day number to ensure it is always a fraction of a second past the hour
        fdoy = doy + hour/24.0 + 1.0e-9
        jd = jd0 + fdoy
        (y,mon,d,h,m,s) = invjday(jd)
        s = int(s)
        fy = y + fdoy/ylen
        fdoypl.append(fdoy)
        pos, vel = satellite.propagate(y,mon,d,h,m,s)
        pole = np.cross(pos,vel)
        polelen = np.linalg.norm(pole)
# calculate sun coordinates
        n = jd - 2451545.0
        L = (280.460 + 0.9856474 * n) % 360
        g = radians((357.528 + 0.9856003 * n) % 360)
        lamda = radians(L + 1.915 * sin(g) + 0.020 * sin(2*g))
        epsilon = radians(23.439 - 0.0000004 * n)
        Xsun = cos(lamda)
        Ysun = cos(epsilon) * sin(lamda)
        Zsun = sin(epsilon) * sin(lamda)
        dotprod = np.dot([Xsun,Ysun,Zsun], pole)
        beta = degrees(asin(dotprod/polelen))
        bpl.append(beta)
        fo.write(oformat % (y, mon, d, h, m, s, fy, pos[0], pos[1], pos[2], beta))
```

```python
rcParams['figure.figsize'] = 12, 8
pylab.ylim(-50, 50)
pylab.xlim(0, 366)
pylab.xlabel('Day of Year')
pylab.ylabel('beta (degrees)', color='b')
pylab.title("Year = "+str(year)) # the `r` before the string indicates a "raw string"
pylab.plot(fdoypl, bpl, label = 'Beta angle')
pylab.plot([-10,400],[0,0],'k')
pylab.plot([-10,400],[45,45],'r')
pylab.plot([-10,400],[-45,-45],'r')
pylab.legend(loc='best',fontsize=12)
#pylab.legend(loc='upper right',fontsize=12)
savefig('beta2015.png')
```
