```python
%matplotlib inline
import time
import calendar
import re
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
plt.rc('figure', figsize = [16, 12])
plt.rc('font', size = 16)
```

```python
f9 = fm.FontProperties(size=7)
f10 = fm.FontProperties(size=10)
```

```python
def month_setup(mon):
    mon.axis('off')
    mon.text(0.0,0.9,'S   M   T   W   T   F   S', fontproperties=f9, color='r')
```

```python
# row and column sharing
f, ((jan, feb, mar, apr), (may, jun, jul, aug), (sep, oct, nov, dec)) = plt.subplots(3, 4, sharex='col', sharey='row')
#jan.text(0.0,0.9,'S   M   T   W   T   F   S', fontproperties=f9, color='r')
jan.text(0.0,0.8,'1 2 3 4 5 6 7', fontproperties=f10, color='g')

jan.set_title('January')
month_setup(jan)
#jan.axis('off')
feb.set_title('February')
month_setup(feb)
#feb.axis('off')
#f.title("%d" % (year))
#f.show()
```
