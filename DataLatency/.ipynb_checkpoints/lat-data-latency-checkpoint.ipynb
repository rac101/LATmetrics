# Cumulative Histograms of LAT data latency times

### Robert Cameron, March 2016

```python
from astropy.table import Table, Column
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import datetime as dt
%matplotlib inline
plt.rc('figure', figsize = [12, 10])
plt.rc('font', size = 16)
```

```python
# read regular ASCII tabular files of data

file1 = "/Users/rc/Documents/GLAST/ISOC/LATmetrics/latency/2016march/s-oct15-mar16.txt"
f1 = Table.read(file1, format = 'ascii')

file2 = "/Users/rc/Documents/GLAST/ISOC/LATmetrics/latency/2016march/s-oct14-oct15.txt"
f2 = Table.read(file2, format = 'ascii')

file3 = "/Users/rc/Documents/GLAST/ISOC/LATmetrics/latency/2016march/s-feb16-mar16.txt"
f3 = Table.read(file3, format = 'ascii')
```

```python
t1 = f1["Total"]
lt1 = np.log10(t1)
t2 = f2["Total"]
lt2 = np.log10(t2)
t3 = f3["Total"]
lt3 = np.log10(t3)
```

```python
n_bins = 100
#n, bins, patches = plt.hist(lt1, n_bins, normed=1, histtype='step', cumulative=True)
plt.hist(lt1, n_bins, normed=1, histtype='step', cumulative=True)
plt.hist(lt2, n_bins, normed=1, histtype='step', cumulative=True)

plt.grid(True)
plt.ylim(0, 1.05)
plt.title('Run Processing Time (hours)')

plt.show()
```

```python
plt.xscale("log")
plt.hist(t1, bins = np.logspace(0, 2, 1000), normed=1, histtype='step', cumulative=True,label='Oct 2015 - Mar 2016')
plt.hist(t2, bins = np.logspace(0, 2, 1000), normed=1, histtype='step', cumulative=True, label='Oct 2014 - Oct 2015')
plt.legend(loc='lower right')
plt.ylim(0, 1.05)
plt.yticks( np.arange(21)/20 )
plt.grid(b=True, which='both', color='0.65',linestyle='-')
plt.title('Run Processing Time (hours)')
plt.ylabel('Normalized Cumulative Distribution')
plt.show()
```

```python
plt.xscale("log")
plt.hist(t3, bins = np.logspace(0, 2, 1000), normed=1, histtype='step', cumulative=True, label='Feb 2016 - Mar 2016')
plt.hist(t1, bins = np.logspace(0, 2, 1000), normed=1, histtype='step', cumulative=True, label='Oct 2015 - Mar 2016')
plt.hist(t2, bins = np.logspace(0, 2, 1000), normed=1, histtype='step', cumulative=True, label='Oct 2014 - Oct 2015')

plt.legend(loc='lower right')
plt.ylim(0, 1.05)
plt.yticks( np.arange(21)/20 )
plt.grid(b=True, which='both', color='0.65',linestyle='-')
plt.title('Run Processing Time (hours)')
plt.ylabel('Normalized Cumulative Distribution')
plt.show()
```

```python
plt.xscale("log")
plt.hist(t3, bins = np.logspace(0, 2, 1000), normed=1, histtype='step', cumulative=True, label='Feb 2016 - Mar 2016')
plt.hist(t2, bins = np.logspace(0, 2, 1000), normed=1, histtype='step', cumulative=True, label='Oct 2014 - Oct 2015')

plt.legend(loc='lower right')
plt.ylim(0, 1.05)
plt.yticks( np.arange(21)/20 )
plt.grid(b=True, which='both', color='0.65',linestyle='-')
plt.title('Run Processing Time (hours)')
plt.ylabel('Normalized\nCumulative\nDistribution')
plt.show()
```
