```python
import matplotlib.pyplot as plt
import numpy as np
import math
from astropy.io import fits

# Define the SAA polygon boundary vertices

lonv = [ 33.9, 24.5, -18.6, -25.7, -36.0, -42.0, -58.8, -93.1, -97.5, -98.5, -92.1, -86.1, 33.9]
latv = [-30.0, -22.6, 2.5, 5.2, 5.2, 4.6, 0.7, -8.6, -9.9, -12.5, -21.7, -30.0, -30.0]
maxmin = [1,1,1,0,-1,-1,-1,-1,-1,-1,-1,0]

maxlon=[-30]*36
minlon=[-30]*36
plat=[90]*36
alat=range(-30,6)

for s in range(12):
    if latv[s] == latv[s+1]:
        slope = 0
    else:
        slope = (lonv[s+1] - lonv[s])/(latv[s+1] - latv[s])
    lat0=min(round(latv[s]),round(latv[s+1]))
    lat1=max(round(latv[s]),round(latv[s+1]))
    print(s,lonv[s],lonv[s+1],latv[s],latv[s+1],lat0,lat1,slope,maxmin[s])
    for l in range(lat0,lat1+1):
        plat = l+30
        # lon = (lat - latv2)*slope + lonv2
        lon = (l - latv[s])*slope + lonv[s]
#        print(s,l,lon)
        if lat0 == lat1:
            maxlon[plat] = max(lonv[s],lonv[s+1],maxlon[plat],lon)
            minlon[plat] = min(lonv[s],lonv[s+1],minlon[plat],lon)
            continue
        if maxmin[s] == 0:    # side of constant latitude
            maxlon[plat] = max(lonv[s],lonv[s+1],maxlon[plat],lon)
            minlon[plat] = min(lonv[s],lonv[s+1],minlon[plat],lon)
        if maxmin[s] == 1:    # set max longitude for latitude l
            maxlon[plat] = max(lon,maxlon[plat])
        if maxmin[s] == -1:   # set min longitude for latitude l
            minlon[plat] = min(lon,minlon[plat])
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
ax.plot(lonv,latv,'o-',color='r')
ax.plot(minlon,alat,'*',color='b')
ax.plot(maxlon,alat,'*',color='g')
ax.grid()
```

```python
mask=(53,361)
mask=np.zeros(mask)
mask1=(53,361)
mask1=np.zeros(mask1)
# for negative maxlon values, we want the more (not less) negative integer longitude
# so inclusive longitude indices into mask array are 
for lat in range(-26,6):
    mlat = lat+26
    plat = lat+30
    i = int(minlon[plat])
    j = math.floor(maxlon[plat])
    i0 = i + 180
    j0 = j + 180
#    print(lat,mlat,plat,alat[plat],minlon[plat],i,maxlon[plat],i,int(maxlon[plat]))
    mask[mlat,i0:j0+1] = 1
# make a 1-longitude-pixel smaller mask image: mask1
    mask[mlat,i0+1:j0] = 1
#mask
```

```python
nlonv=np.array(lonv)+180
nlatv=np.array(latv)+26
fig, ax = plt.subplots(figsize=(16, 8))
ax.imshow(mask,origin='lower')
ax.plot(nlonv,nlatv,color='r')
plt.ylim(-10,40)
plt.xlim(70,220)
#mask
```
