```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import DatetimeTickFormatter
#from matplotlib.dates import (MONTHLY, YEARLY, DateFormatter, rrulewrapper, RRuleLocator, drange)
```

```python
url = "http://www.slac.stanford.edu/exp/glast/ops/notebook/ssr.txt"
table = pd.read_csv(url,names=('date','gbps'),parse_dates=[0],sep=' ')
dnt = table['date']
gbps = table['gbps']
output_notebook()
#table.describe
```

```python
# create a new plot with a datetime axis type
p = figure(width=800, height=350, x_axis_type="datetime",y_range=(100,160))
#p = figure(width=800, height=350,y_range=(100,160),x_minor_ticks=9)
#p.xaxis.desired_num_ticks=9
p.line(dnt, gbps, color='navy')
p.xaxis.formatter = DatetimeTickFormatter(seconds=["%M:%S"],
                                            minutes=["%M:%S"],
                                            minsec=["%M:%S"],
                                            hours=["%M:%S"],years=["%Y"]
                                            )
```

```python
# NEW: customize by setting attributes

p.grid.grid_line_alpha = 1
p.yaxis.axis_label = 'SSR Usage: Gbits/day'
p.xgrid.minor_grid_line_alpha = 1

# show the results
show(p)
```
