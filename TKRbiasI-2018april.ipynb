```python
# read the csv data file

cols = ['d','s','t0','t1','t2','t3','t4','t5','t6','t7','t8','t9','ta','tb','tc','td','te','tf']
file = "/Users/rac/Documents/GLAST/ISOC/LATmetrics/10yearLAT/april.txt"
f = pd.read_csv(file,names=cols,parse_dates=['d'],skiprows=1)
d = f['d']
t0 = f['t0']
t1 = f['t1']
t2 = f['t2']
t3 = f['t3']
t4 = f['t4']
t5 = f['t5']
t6 = f['t6']
t7 = f['t7']
t8 = f['t8']
t9 = f['t9']
ta = f['ta']
tb = f['tb']
tc = f['tc']
td = f['td']
te = f['te']
tf = f['tf']

t0 -= 0.182495306803
t1 -= 0.182150759042
t2 -= 0.180980068968
t3 -= 0.178817817555
t4 -= 0.293007336276
t5 -= 0.18078428991
t6 -= 0.225185115969
t7 -= 0.152437598727
t8 -= 0.2284879803
t9 -= 0.179055345909
ta -= 0.168437627654
tb -= 0.15360937061
tc -= 0.198550999442
td -= 0.178332619201
te -= 0.133621645159
tf -= 0.166996182669

#t0 = f['t0'] - 0.182495306803
#t1 = f['t1'] - 0.182150759042
#t2 = f['t2'] - 0.180980068968
#t3 = f['t3'] - 0.178817817555
#t4 = f['t4'] - 0.293007336276
#t5 = f['t5'] - 0.18078428991
#t6 = f['t6'] - 0.225185115969
#t7 = f['t7'] - 0.152437598727
#t8 = f['t8'] - 0.2284879803
#t9 = f['t9'] - 0.179055345909
#ta = f['ta'] - 0.168437627654
#tb = f['tb'] - 0.15360937061
#tc = f['tc'] - 0.198550999442
#td = f['td'] - 0.178332619201
#te = f['te'] - 0.133621645159
#tf = f['tf'] - 0.166996182669
```

```python
# apply some data smoothing 

smoothlength = 99

st0 = t0.rolling(window=smoothlength).mean()
st1 = t1.rolling(window=smoothlength).mean()
st2 = t2.rolling(window=smoothlength).mean()
st3 = t3.rolling(window=smoothlength).mean()
st4 = t4.rolling(window=smoothlength).mean()
st5 = t5.rolling(window=smoothlength).mean()
st6 = t6.rolling(window=smoothlength).mean()
st7 = t7.rolling(window=smoothlength).mean()
st8 = t8.rolling(window=smoothlength).mean()
st9 = t9.rolling(window=smoothlength).mean()
sta = ta.rolling(window=smoothlength).mean()
stb = tb.rolling(window=smoothlength).mean()
stc = tc.rolling(window=smoothlength).mean()
std = td.rolling(window=smoothlength).mean()
ste = te.rolling(window=smoothlength).mean()
stf = tf.rolling(window=smoothlength).mean()
```

```python
fig = plt.figure(figsize = [12,9])
ax = fig.add_subplot(111)
ax.set_ylim(0.1,0.8)
ax.set_xlim([dt.date(2018, 4, 1), dt.date(2018, 4, 29)])

ax.plot(d,st0,label='TKR 0',color='g')
ax.plot(d,st1,label='TKR 1',color='b')
ax.plot(d,st2,label='TKR 2',color='r')
ax.plot(d,st3,label='TKR 3',color='c')
ax.plot(d,st4,label='TKR 4',color='m')
ax.plot(d,st5,label='TKR 5',color='k')

ax.plot(d,st6,label='TKR 6',color='g')
ax.plot(d,st7,label='TKR 7',color='b')
ax.plot(d,st8,label='TKR 8',color='r')
ax.plot(d,st9,label='TKR 9',color='c')
ax.plot(d,sta,label='TKR A',color='m')
ax.plot(d,stb,label='TKR B',color='k')

ax.plot(d,stc,label='TKR C',color='g')
ax.plot(d,std,label='TKR D',color='b')
ax.plot(d,ste,label='TKR E',color='r')
ax.plot(d,stf,label='TKR F',color='c')

ax.legend(loc='lower right')
ax.grid(True)
#plt.xlabel("Time")
plt.ylabel("Bias Current (mA)",fontsize=15)
plt.title("Tracker Towers Bias Currents recovery after LAT shutdown in March 2018",fontsize=15)

# set Major and Minor ticks on the time axis


plt.gca().xaxis.set_major_locator(mdates.DayLocator((1,8,15,22,29)))
#plt.gca().xaxis.set_major_locator(mdates.DayLocator((0,5,10,15,20,25,30)))
#plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.gca().xaxis.set_minor_locator(mdates.DayLocator(interval=1))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%b-%d"))
plt.gcf().autofmt_xdate()

plt.show()
```
