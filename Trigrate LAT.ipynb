```python
#from astropy.table import Table, Column
#import numpy as np
import matplotlib.pyplot as plt
#import matplotlib.dates as mdates
import pandas as pd
import datetime as dt
#import re
%matplotlib notebook
from matplotlib.dates import (MONTHLY, YEARLY, DateFormatter, rrulewrapper, RRuleLocator, drange)
```

```python
url = "http://www.slac.stanford.edu/exp/glast/ops/notebook/trigrate.txt"
table = pd.read_csv(url,names=('date','trigs'),parse_dates=[0],sep=' ')
dnt = table['date']
trigs = table['trigs']
```

```python
plt.figure(figsize = [15, 8])
plt.rc('font', size = 15)
fig, ax1 = plt.subplots(1, figsize = [15, 8])

yrule = rrulewrapper(YEARLY, interval=1)
yloc = RRuleLocator(yrule)
mrule = rrulewrapper(MONTHLY, interval=3)
mloc = RRuleLocator(mrule)

ax1.xaxis.set_major_locator(yloc)
ax1.xaxis.set_minor_locator(mloc)

# set plot date range
dntmin="2012-01-01"
dntmax="2019-01-01"
dnt0 = dt.datetime.strptime(dntmin,"%Y-%m-%d")
dnt1 = dt.datetime.strptime(dntmax,"%Y-%m-%d")
plt.xlim(dnt0,dnt1)
plt.plot(dnt, trigs, 'r')
plt.grid(True)
plt.ylim(1500, 2200)
plt.ylabel('LAT Event Triggers/second')
```
