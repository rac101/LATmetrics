```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
EarthRadius = 6378.0

SemiMajor = np.arange(1,56400)
Alt = SemiMajor - EarthRadius

GM = 398600.44;  # km^3 s^-2 (Wertz 1978)

Period = 2*np.pi*np.sqrt(SemiMajor**3/GM)
RevsPerDay = 86400/Period
OrbitLength = SemiMajor*np.pi*2
Speed = OrbitLength/Period

c = 299792.458
csq= c*c
r_Schwartzschild = 2*GM/csq
r_S_mm = r_Schwartzschild * 1e6

rotating = 1.5 # scale factor = 1.0 for non-rotating system or 1.5 for rotating system

Dilation = np.sqrt(1-(Speed/c)**2)
Redshift = np.sqrt(1 - rotating*r_Schwartzschild/SemiMajor) # https://en.wikipedia.org/wiki/Gravitational_time_dilation
Redshift0 = np.sqrt(1 - rotating*r_Schwartzschild/EarthRadius)

ddt = (Dilation-1) * 1e9
drt = (Redshift-Redshift0) * 1e9
dtot = ddt + drt

i = 46377
SemiMajor[i],Alt[i],Period[i],RevsPerDay[i],OrbitLength[i],Speed[i],Dilation[i],ddt[i],Redshift[i],drt[i],r_S_mm
```

```python
plt.figure(figsize = [10, 5])
plt.rc('font', size = 14)

plt.xlabel('Orbit Altitude (km)')
plt.ylabel('Nano second per second')
plt.title('Time gained or lost by Satellite clock relative to clock used by Earth Observer',size=14)

plt.minorticks_on()
plt.grid(which='major')

plt.plot(Alt[6377:],ddt[6377:],label='Time Dilation')
plt.plot(Alt[6377:],drt[6377:],label='Grav Redshift')
plt.plot(Alt[6377:],dtot[6377:],label='Net time change')
plt.xlim(0,50000)
plt.legend()
#plt.ylim(-1000,1000)
```
