```python
fig, ax = plt.subplots()
ax.plot([3,1,2,4,0,5,3,2,0,2,4],"o")
plt.close(fig)

vline = ax.axvline(1)
hline = ax.axhline(0.5)

def set_cursor(x, y):
    vline.set_xdata((x, x))
    hline.set_ydata((y, y))
    display(fig)

interact(set_cursor, x=(1, 9, 0.01), y=(0, 5, 0.01))
```

```python
x = np.linspace(0, 40, 100)
y = np.sin(x)*5

def set_range(imin, imax):
    plt.ylim(-6, 6)
    plt.xlim(0, 40)
    plt.plot(x[imin:imax], y[imin:imax], '.-')
    plt.title('Min='+str(imin)+'     Max='+str(imax))
#set_range(0,100)
interact(set_range, imin=(0, 20), imax=(80, 100))
```

```python
x = np.linspace(0, 40, 200)
y = np.sin(x)*5

fig, ax = plt.subplots()
ax.plot(x,y,".-")
plt.close(fig)

vline = ax.axvline(1)

def set_cursor(idx):
    vline.set_xdata((x[idx], x[idx]))
    display(fig)
    xx = x[idx]
    yy = y[idx]
    return xx, yy

interact((xxx,yyy)=set_cursor, idx=(0, 199))
def h(idx=55):
    print(idx, xxx, yyy)
```
