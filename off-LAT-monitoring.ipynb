```python
from astropy.table import Table, Column
from astropy.time import Time
import numpy as np
from os import walk
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
tutc18 = Time('2018-01-01 00:00:00')
```

```python
path = "/Users/rac/Documents/GLAST/ISOC/LATmetrics/off-LAT/csv2/"

files = []
for (dirpath, dirnames, filenames) in walk(path):
    files.extend(filenames)
#files
```

```python
plt.figure(figsize = [16, 9])
plt.rc('font', size = 18)
for f in files:
    t = Table.read(path+f, format='ascii')
    times = t['TIME']
    tutc = Time(times)-tutc18
    names = t.colnames
    plt.figure(figsize = [16, 9])
    plt.rc('font', size = 18)
    plt.title(f)
    plt.xlabel('Day in Year 2018')
    for n in names[2:]:
        plt.plot(tutc.jd,t[n],'.',label=n)
        plt.legend(bbox_to_anchor=(1.02,1),loc=2,borderaxespad=0.)
    plt.show()
```
